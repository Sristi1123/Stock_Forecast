```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
df = pd.read_csv('/content/Apple.csv', parse_dates=['Date'])
df.set_index('Date', inplace=True)
df = df.sort_index()
df = df[['Close']].dropna()
df = df.asfreq('D')
df['Close'] = df['Close'].interpolate('linear')
df.dropna(subset=['Close'], inplace=True)
```

```python
df_prophet = df['Close'].reset_index()
df_prophet.columns = ['ds', 'y']
train_prophet = df_prophet.iloc[:-30]
test_prophet = df_prophet.iloc[-30:]
```

```python
model_prophet = Prophet()
model_prophet.fit(train_prophet)
```

```python
future = model_prophet.make_future_dataframe(periods=30)
forecast = model_prophet.predict(future)
model_prophet.plot(forecast)
model_prophet.plot_components(forecast)
plt.show()
```

```python
predicted = forecast[['ds', 'yhat']].tail(30)
actual = test_prophet.copy()
merged = predicted.merge(actual, on='ds')
merged.columns = ['Date', 'Predicted', 'Actual']
```

```python
plt.figure(figsize=(10,5))
plt.plot(merged['Date'], merged['Actual'], label='Actual')
plt.plot(merged['Date'], merged['Predicted'], label='Predicted', linestyle='--')
plt.legend()
plt.title("Actual vs Predicted - Prophet")
plt.grid(True)
plt.show()
```

```python
prophet_rmse = np.sqrt(mean_squared_error(merged['Actual'], merged['Predicted']))
print(f"Prophet RMSE: {prophet_rmse:.4f}")
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv("prophet_forecast.csv", index=False)

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
```

```python
def create_sequences(data, seq_len=60):
    X, y = [], []
    for i in range(seq_len, len(data)):
        X.append(data[i - seq_len:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

X, y = create_sequences(df_scaled, seq_len=60)
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(X.shape[1], 1)))
model_lstm.add(LSTM(50))
model_lstm.add(Dense(1))
model_lstm.compile(optimizer='adam', loss='mean_squared_error')
model_lstm.fit(X, y, epochs=10, batch_size=32)
```

```python
predicted_full = model_lstm.predict(X)
predicted_full = scaler.inverse_transform(predicted_full)
actual_full = scaler.inverse_transform(y.reshape(-1, 1))
```

```python
plt.figure(figsize=(12,6))
plt.plot(actual_full, label='Actual')
plt.plot(predicted_full, label='LSTM Predicted')
plt.title('LSTM - Actual vs Predicted Closing Price')
plt.xlabel("Days")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
forecast_input = df_scaled[-60:].reshape(1, 60, 1)
future_preds = []

for _ in range(30):
    pred = model_lstm.predict(forecast_input)[0][0]
    future_preds.append(pred)
    forecast_input = np.append(forecast_input[:, 1:, :], [[[pred]]], axis=1)
```

```python
future_preds_actual = scaler.inverse_transform(np.array(future_preds).reshape(-1, 1))
last_date = df.index[-1]
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Historical Close')
plt.plot(future_dates, future_preds_actual, label='LSTM Forecast (Next 30 Days)', color='orange')
plt.title("LSTM Forecast - Next 30 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
```

```python
# ADF Test for Stationarity
result = adfuller(df['Close'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])
if result[1] <= 0.05:
    print(" Data is stationary")
else:
    print(" Data is NOT stationary")
```

```python
df_diff = df['Close'].diff().dropna()
# ARIMA model (example with p=5, d=1, q=2)
arima_model = ARIMA(df['Close'], order=(5, 1, 2))
arima_result = arima_model.fit()
print(arima_result.summary())
```

```python
# Forecast for next 30 days
arima_forecast = arima_result.forecast(steps=30)

# Create date range for forecast
arima_dates = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=30)

# Plot forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Historical')
plt.plot(arima_dates, arima_forecast, label='ARIMA Forecast (30 Days)', color='red')
plt.title("ARIMA Forecast - Next 30 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Seasonal ARIMA example: (1,1,1)x(1,1,1,12)
sarima_model = SARIMAX(df['Close'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 7))
sarima_result = sarima_model.fit()
print(sarima_result.summary())
```

```python
arima_result.plot_diagnostics(figsize=(15, 10))
plt.show()
```

```python
sarima_forecast = sarima_result.get_forecast(steps=30)
sarima_pred = sarima_forecast.predicted_mean
sarima_ci = sarima_forecast.conf_int()

# Plot SARIMA forecast
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Historical')
plt.plot(arima_dates, sarima_pred, label='SARIMA Forecast (30 Days)', color='purple')
plt.fill_between(arima_dates, sarima_ci.iloc[:, 0], sarima_ci.iloc[:, 1], color='purple', alpha=0.2)
plt.title("SARIMA Forecast - Next 30 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Prepare Prophet forecast
prophet_forecast = forecast[['ds', 'yhat']].tail(30)
prophet_forecast.columns = ['Date', 'Prophet']

# Prepare LSTM forecast
lstm_forecast = pd.DataFrame({'Date': future_dates, 'LSTM': future_preds_actual.flatten()})

# Reuse ARIMA and SARIMA forecasts
arima_df = pd.DataFrame({'Date': arima_dates, 'ARIMA': arima_forecast})
sarima_df = pd.DataFrame({'Date': arima_dates, 'SARIMA': sarima_pred.values})

# Merge all into one
compare_df = prophet_forecast.merge(lstm_forecast, on='Date')
compare_df = compare_df.merge(arima_df, on='Date')
compare_df = compare_df.merge(sarima_df, on='Date')
compare_df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(16, 6))
plt.plot(df.index[-90:], df['Close'][-90:], label='Historical (Last 90 Days)', color='black', linestyle='--')
plt.plot(compare_df.index, compare_df['Prophet'], label='Prophet', marker='o')
plt.plot(compare_df.index, compare_df['LSTM'], label='LSTM', marker='^')
plt.plot(compare_df.index, compare_df['ARIMA'], label='ARIMA', marker='s')
plt.plot(compare_df.index, compare_df['SARIMA'], label='SARIMA', marker='d')

plt.title(" 30-Day Forecast Comparison: Prophet vs LSTM vs ARIMA vs SARIMA")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Prophet RMSE (already calculated)
prophet_rmse = np.sqrt(mean_squared_error(merged['Actual'], merged['Predicted']))
print(f"Prophet RMSE: {prophet_rmse:.4f}")
```

```python
# True values (last 30 actual Close prices)
actual_30 = df['Close'].tail(30).values
lstm_rmse = np.sqrt(mean_squared_error(actual_30, future_preds_actual))
print(f"LSTM RMSE: {lstm_rmse:.4f}")
```

```python
# Actual and predicted for same period
actual_arima = df['Close'].tail(30).values
arima_rmse = np.sqrt(mean_squared_error(actual_arima, arima_forecast))
print(f"ARIMA RMSE: {arima_rmse:.4f}")
```

```python
# Actual and SARIMA predicted
sarima_rmse = np.sqrt(mean_squared_error(actual_arima, sarima_pred))
print(f"SARIMA RMSE: {sarima_rmse:.4f}")
```

```python
# RMSE summary
rmse_scores = {
    'Prophet': prophet_rmse,
    'LSTM': lstm_rmse,
    'ARIMA': arima_rmse,
    'SARIMA': sarima_rmse
}

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(rmse_scores.keys(), rmse_scores.values(), color=['blue', 'orange', 'red', 'purple'])
plt.title(" RMSE Comparison of Models (Lower = Better)")
plt.ylabel("RMSE")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
rmse_df = pd.DataFrame(list(rmse_scores.items()), columns=['Model', 'RMSE'])
rmse_df.to_csv("model_rmse_comparison.csv", index=False)
```
